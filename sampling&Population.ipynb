```python
dataset = pd.read_csv('Dataset.csv')
print(dataset.shape)
print(dataset.describe())
print(dataset.head(5))
```

Simple Random Sampling

```python
simpleRandomSample=dataset.sample(n=5).sort_values(by='Id')
mean_simpleRandomSample=round(simpleRandomSample['SalePrice'].mean(),3)
print(mean_simpleRandomSample)
print(simpleRandomSample)
```

Systematic Sampling

```python
index=np.arange(0,len(dataset),step=5) #np is numpy
#print(index)
systematicSample=dataset.iloc[index]
mean_systematicSample=round(systematicSample['SalePrice'].mean(),3)
print("Mean of Systematic Sample:",mean_simpleRandomSample)
print(mean_systematicSample)
```

Cluster Sampling

```python
#grouping the data into clusters based on some parameter
n=5 # how many clusters we are going to take
#divide the units into clusers of equal size
dataset['cluster_id']=np.repeat([range(1,n+1)],len(dataset)/n) #range of 1,6(n+1) and length of dataset/2
index=[]
# For this formula,cluster id must be an even number
for i in range(0,len(dataset)):
  if dataset['cluster_id'].iloc[i]%2==0:
    index.append(i)
clusterSample=dataset.iloc[index]
mean_clusterSample=round(clusterSample['SalePrice'].mean(),3)
print("Mean of Cluster Sample:",mean_clusterSample)
print(clusterSample)
```

Stratified Random Sampling

```python
#dividing data into subgroups
#taking random samples from the subgroups
#divide the data into sub categories
#strata =1 ,2 anything
dataset=pd.read_csv('Dataset.csv')
dataset['strata']=np.repeat([1,2],len(dataset)/2)
index=[]
for i in range(0,len(dataset)):
    index.append(i)
stratifiedRandomSample=dataset.iloc[index]
stratifiedRandomSample
```

```python
from sklearn.model_selection import StratifiedShuffleSplit as sss
#setting the split criteria
split=sss(n_splits=1,test_size=8)
#perform data frame split
for x,y in split.split(dataset,dataset['strata']):
  stratifiedRandomSample=dataset.iloc[y].sort_values(by='SalePrice')
stratifiedRandomSample
```
